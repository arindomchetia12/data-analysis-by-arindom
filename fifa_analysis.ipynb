Now i'm going to add another column showing us outcome of the match. i will use a conditianal function which compares two columns and return a specific string, and then apply that function in the creation of a new column

```python
def outcome(matches_df):
    if matches_df['Home_Team_Goals'] > matches_df['Away_Team_Goals']:
        return 'Home_Team_Win'
    if matches_df['Home_Team_Goals'] < matches_df['Away_Team_Goals']:
        return 'Away_Team_Win'
    return 'DRAW'

matches_df['outcome_of_the_match'] = matches_df.apply(lambda x: outcome(x), axis=1)
```

TODO - DISTRIBUTION OF HOME TEAM GEALS AND AWAY TEAM GOALS

```python
plt.figure(figsize=(12,13))
plt.subplot(211)
sns.distplot(matches_df["Home_Team_Goals"],color="black",rug=True)
plt.xticks(np.arange(0,12,1))
plt.title("Distribution of Home Team Goals",color='g')


plt.subplot(212)
sns.distplot(matches_df["Away_Team_Goals"],color="g",rug=True)
plt.xticks(np.arange(0,12,1))
plt.title("Distribution of Away Team Goals",color='g')
plt.show()
```

the above graphs show that scores are in range of 0 and 4 goals, and it very rare for a team weather home tome or away team to score more than 4 goals.

**TODO** - **AVERAGE NUMBER OF ATTENDANTS PER YEAR**

```python
attendance_df = matches_df.groupby("Year")["Attendance"].mean().reset_index()
attendance_df
```

```python
attendance_df["Year"] = attendance_df["Year"].astype(int)
# Pass the data as a keyword argument using the 'data' parameter
attend = sns.pointplot(x="Year", y="Attendance", data=attendance_df, color="w")
attend.set_facecolor("k")
plt.grid(True,color="yellow",alpha=.5)
plt.title("Average attendence by year",color='g')
plt.show()
```
